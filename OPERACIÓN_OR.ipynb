OPERACIÓN OR

```python
# NEURONA McCullock-Pitts - OPERACIÓN OR
# ======================================

# Configuración de la neurona para OR
print("=== CONFIGURACIÓN NEURONA OR ===")
print("Pesos: w1 = 1, w2 = 1")
print("Umbral: θ = 1")
print("Función de activación: Escalón")

def neurona_OR(x1, x2):
    # Parámetros de la neurona
    w1 = 1
    w2 = 1
    theta = 1

    # Calcula la suma ponderada
    suma = (w1 * x1) + (w2 * x2)

    # Aplica función de activación escalón
    if suma >= theta:
        return 1
    else:
        return 0

# Tabla de verdad OR
print("\n=== TABLA DE VERDAD OR ===")
print("x1 | x2 | Salida OR")
print("---+----+----------")

# Probamos todas las combinaciones
entradas = [(0, 0), (0, 1), (1, 0), (1, 1)]

for x1, x2 in entradas:
    salida = neurona_OR(x1, x2)
    print(f" {x1} |  {x2} |     {salida}")

# Verificación detallada
print("\n=== VERIFICACIÓN DETALLADA ===")
for x1, x2 in entradas:
    suma = (1 * x1) + (1 * x2)
    salida = neurona_OR(x1, x2)
    print(f"Entradas: ({x1}, {x2})")
    print(f"Suma ponderada: {1}*{x1} + {1}*{x2} = {suma}")
    print(f"Comparación con θ={1}: {suma} >= {1}? {suma >= 1}")
    print(f"Salida: {salida}")
    print("-" * 40)
```

¿QUÉ ESTÁ HACIENDO LA NEURONA?

La neurona OR funciona como un “detector de presencia”. Su lógica es bastante sencilla pero muy práctica. Recibe dos señales de entrada, x1 y x2, que pueden ser 0 o 1. Cada una se multiplica por su peso correspondiente, que en este caso es 1 para ambas. Luego se suman esos valores y el resultado se compara con un umbral, que es θ = 1.
Si la suma es mayor o igual que ese umbral, la neurona se activa y produce una salida de 1. Si la suma no alcanza el umbral, la neurona permanece inactiva y su salida es 0.

En pocas palabras, la neurona OR “se enciende” cuando al menos una de sus entradas recibe una señal, y solo se mantiene “apagada” cuando las dos están en cero. Es como una alarma que se activa con que cualquiera de sus sensores detecte movimiento.
