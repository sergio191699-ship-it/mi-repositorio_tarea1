OPERACIÓN NOT

```python
# NEURONA McCullock-Pitts - OPERACIÓN NOT
# =======================================

# Configuración de la neurona para NOT
print("=== CONFIGURACIÓN NEURONA NOT ===")
print("Peso: w1 = -1")
print("Umbral: θ = 0")
print("Función de activación: Escalón")

def neurona_NOT(x):
    # Parámetros de la neurona
    w1 = -1
    theta = 0

    # Calcula la suma ponderada
    suma = w1 * x

    # Aplica función de activación escalón
    if suma >= theta:
        return 1
    else:
        return 0

# Tabla de verdad NOT
print("\n=== TABLA DE VERDAD NOT ===")
print("x | Salida NOT")
print("--+-----------")

# Probamos ambas entradas posibles
entradas = [0, 1]

for x in entradas:
    salida = neurona_NOT(x)
    print(f" {x} |     {salida}")

# Verificación detallada
print("\n=== VERIFICACIÓN DETALLADA ===")
for x in entradas:
    suma = (-1) * x
    salida = neurona_NOT(x)
    print(f"Entrada: {x}")
    print(f"Suma ponderada: {-1}*{x} = {suma}")
    print(f"Comparación con θ={0}: {suma} >= {0}? {suma >= 0}")
    print(f"Salida: {salida}")
    print("-" * 40)
```

¿QUÉ ESTÁ HACIENDO LA NEURONA?

La neurona NOT actúa como un “inversor lógico”, es decir, toma una señal y devuelve lo contrario. Su funcionamiento es sencillo: recibe una sola entrada, que puede ser 0 o 1, y la multiplica por un peso negativo (w1 = –1). Ese peso se encarga de invertir el valor. Luego compara el resultado con un umbral igual a 0.
Si el valor obtenido es mayor o igual que el umbral, la neurona se activa y la salida es 1. Si el resultado queda por debajo, la neurona se mantiene inactiva y la salida es 0.

En pocas palabras, la neurona NOT siempre responde con lo opuesto a lo que recibe. Si le das un 1, devuelve un 0; si le das un 0, devuelve un 1.

Es como un interruptor invertido de luz: cuando presionas el botón (entrada 1) la luz se apaga (salida 0), y cuando no presionas nada (entrada 0) la luz permanece encendida (salida 1).

Este tipo de neurona es clave para construir operaciones lógicas más complejas y muestra cómo incluso la negación puede implementarse con un modelo tan simple.
